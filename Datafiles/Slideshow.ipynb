## Breed Analysis
In the first section, we are going to take a look at each the different rated breed to find out which breed is the highest rated as well as the most rated. Since there are so many options, we're going to limit the cut off at least 20 counts.

### Most Rated Breeds
The visualization shows us that the golden retriever is the most rated dog on the Weratedogs twitter profile. 

```python
top_breeds=viz.groupby('breed').filter(lambda x: len(x) >= 20)
top_breeds['breed'].sort_values(ascending = True).value_counts().plot(kind = 'barh')
plt.xlabel('Count')
plt.ylabel('Breed')
```
